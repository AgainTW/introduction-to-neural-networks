## 創建DataFrame與Series

### 建立 Series

* **透過 list 建立 Series**

```python
height = [168,175,173,180,192,178]
name = ['olivia','harry','jhon','mike','josh','leo']
a = pd.Series(data=height) # 只顯示資料
b = pd.Series(data=height,index=name) # 設定index的名稱

print('===== a ======')
print(a)
print('===== b ======')
print(b)
```

* **透過array建立一個Series**

```python
height = np.array([168,175,173,180,192,178])
name = ['olivia','harry','jhon','mike','josh','leo']
a = pd.Series(data=height) # 只顯示資料
b = pd.Series(data=height,index=name) # 設定index的名稱

print('===== a ======')
print(a)
print('===== b ======')
print(b)
```

* **透過dict建立Series**

```python
height = {'olivia':168,'harry':175,'jhon':173,'mike':180,'josh':192,'leo':178}
name = ['olivia','harry','jhon','mike','josh','leo']
a = pd.Series(data=height) # 只顯示資料
b = pd.Series(data=height,index=name) # 設定index的名稱

print('===== a ======')
print(a)
print('===== b ======')
print(b)
```

### 怎麼處理空白值?

```python
# 缺失資料讀取
df_loss = pd.read_excel('./小番茄價格/小番茄-嬌女(缺失).xlsx')
```

```python
df_loss.isna().sum()
```

```python
df_loss[df_loss['Avg_Price'].isna()]
```

### 替換/刪除資料

* **利用loc來替換資料**

```python
# 將 df_loss_fill的第1筆資料的平均售價換成當日最高價與最低價的平均值
upper = df_loss_fill.loc[15,'Upper_Price']
lower = df_loss_fill.loc[15,'Lower_Price']
df_loss_fill.loc[15,'Avg_Price'] = (upper+lower)/2
df_loss_fill.loc[10:15]
```

* **利用 replace 來批量替換資料**

```python
# 把所有資料的NULL替換成20
upper = df_loss_fill.loc[13,'Upper_Price']
lower = df_loss_fill.loc[13,'Lower_Price']
df_loss_fill20 = df_loss_fill.replace('NULL',20)
df_loss_fill20.loc[10:15]
```

* **drop(index) - 刪除資料**

```python
# 刪除平均售價是NULL的資料
df_drop = df_loss.drop(df_loss[df_loss['Avg_Price'].isna()].index)
print(df_drop.isna().sum())
df_drop
```

### 讀取多張工作表的 excel 檔

* **利用dict的方式來讀取**

```python
df_dict = pd.read_excel('./小番茄價格/番茄價格(工作表).xlsx',sheet_name=None)
sheet_name_list = list(df_dict.keys())
total_df = pd.DataFrame()
for name in sheet_name_list:
    data = df_dict[name]
    total_df = pd.concat([total_df,data],ignore_index=True)
total_df
```

* **利用ExcelFile與read_excel的方式讀取**

```python
xls = pd.ExcelFile('./小番茄價格/番茄價格(工作表).xlsx')
sheet_name_list = xls.sheet_names
total_df = pd.DataFrame()
for name in sheet_name_list:
    data = pd.read_excel(xls,name)
    total_df = pd.concat([total_df,data],ignore_index=True)
total_df
```

* **利用ExcelFile與parse的方式讀取**

```python
xls = pd.ExcelFile('./小番茄價格/番茄價格(工作表).xlsx')
sheet_name_list = xls.sheet_names
sheet_length = len(sheet_name_list)
total_df = pd.DataFrame()
for i in range(sheet_length):
    data = xls.parse(i)
    total_df = pd.concat([total_df,data],ignore_index=True)
total_df
```

### 圖形輸出(matplolib)

* **長條圖**

```python
df = pd.read_excel('./小番茄價格/小番茄-聖女.xlsx')
market_name = list(set(df['MarketName']))
quantity_list = list()
for name in market_name:
    trans_q = df[df['MarketName']==name]['Trans_Quantity'].sum()
    quantity_list.append(trans_q)

plt.figure(figsize=(12,6))
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.title('聖女小番茄交易分佈',fontsize=16)
plt.xlabel('市場',fontsize=16)
plt.ylabel('交易量',fontsize=16)
plt.xticks(fontsize=14)
plt.bar(market_name,quantity_list)
plt.show()
```

* **圓餅圖**

```python
# 以圓餅圖呈現小番茄的市場占比
df = pd.read_excel('./小番茄價格/小番茄-聖女.xlsx')
market_name = list(set(df['MarketName']))
quantity_list = list()
for name in market_name:
    trans_q = df[df['MarketName']==name]['Trans_Quantity'].sum()
    quantity_list.append(trans_q)
quantity_list = list(map(int,quantity_list))


plt.figure(figsize=(10,10))
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.title('市場占比',fontsize=14)
plt.pie(quantity_list , labels=market_name,autopct='%1.1f%%',textprops={'fontsize':14})
plt.show()
```

* **折線圖**

```python
import matplotlib.ticker as mticker

market_name = '台北一'
df = pd.read_excel('./小番茄價格/小番茄-聖女.xlsx')
price = df[df['MarketName']==market_name]['Avg_Price']
date = df[df['MarketName']==market_name]['TransDate']

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(20,6))
ax=plt.subplot()
plt.title(market_name)
ax.plot(date,price)
tick_spacing = len(date)/15
ax.xaxis.set_major_locator(mticker.MultipleLocator(tick_spacing))
plt.show()
```
